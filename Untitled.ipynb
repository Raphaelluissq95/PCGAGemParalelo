```python
class ElementIndividual():
    def __init__(self, max_ground, max_ground_sequence, max_ground_height, min_height_block, max_block, block_type, block_seq, max_enemy, enemy_type, max_coin, max_coin_height, chromossomeSize):
        self.max_coin = int(max_coin)
        self.max_coin_height = int(max_coin_height)
        self.max_enemy = int(max_enemy)
        self.enemy_type = int(enemy_type)
        self.min_height_block = int(min_height_block)
        self.max_block = int(max_block)
        self.block_type = int(block_type)
        self.max_ground = int(max_ground)
        self.max_ground_sequence = int(max_ground_sequence)
        self.max_ground_height = int(max_ground_height)
        self.chromossomeSize = int(chromossomeSize)
        self.block_seq = int(block_seq)

        self.g = ['-', 'X', '|', '%'] # ground
        self.b = ['-', '#', '?', '@', '1', '2', 'D', 'S', 'C', 'U', 'L'] # blocks
        self.e = ['-', 'y', 'E', 'g', 'k', 'r'] # enemies
        self.c = ['-', 'o'] # coins

    def _groundIndividual(self):
        # gound parts (gparts) 1 - 4
        # ground size (gsize) 1 - 132
        # ground maximum height (ghmax) 0 - 4

        g_vector = []
        sequence = 0
        max_sequence = 0
        height = 0
        while sequence < self.chromossomeSize:
            max_sequence = randint(1, int(self.max_ground_sequence))
            if sequence + max_sequence <= self.chromossomeSize:
                height = randint(0, self.max_ground_height)
                for _ in range(max_sequence):
                    g_vector.append(height)
                sequence += max_sequence

        return g_vector

    def _blockIndividual(self):
        # block maximum height (bhmax) 1 - 4
        # block maximum (bmax) 1 - 132
        # block type (btype) 1 - 11

        b_vector = []
        sequence = 0
        max_sequence = 0
        type = 0
        while sequence < self.chromossomeSize:
            max_sequence = randint(1, int(self.block_seq))
            if sequence + max_sequence <= self.chromossomeSize:
                type = randint(1, self.block_type)
                for _ in range(max_sequence):
                    b_vector.append(type)
                sequence += max_sequence
            if sequence + (self.block_seq - max_sequence) <= self.chromossomeSize:
                for _ in range(self.block_seq - max_sequence):
                    b_vector.append(0)
                sequence += (self.block_seq - max_sequence)
            
        if len(b_vector) < 132:
            falta = 132 - len(b_vector)
            for _ in range(0, falta):
                b_vector.append(0)

        return b_vector

    def _enemyIndividual(self):
        # enemy maximum (emax) 1 - 132
        # enemy type (etype) 1 - 9

        e_vector = []
        enemy = self.max_enemy
        for _ in range(0, int(enemy)):
            for _ in range(0, int((self.chromossomeSize/self.max_enemy) - 1)):
                e_vector.append(0)
            if enemy > 0:
                e_vector.append(randint(1, self.enemy_type))
                enemy -= 1

        if len(e_vector) < 132:
            falta = 132 - len(e_vector)
            for _ in range(0, falta):
                e_vector.append(0)
        
        return e_vector

    def _coinIndividual(self):
        # coin maximum (cmax) 1 - 132
        # coin maximum height (chmax) 1 - 4

        c_vector = []
        coin = self.max_coin
        for _ in range(0, int(coin)):
            for _ in range(0, int((self.chromossomeSize/self.max_coin) - 1)):
                c_vector.append(0)
            if coin > 0:
                c_vector.append(1)
                coin -= 1

        if len(c_vector) < 132:
            falta = 132 - len(c_vector)
            for _ in range(0, falta):
                c_vector.append(0)
        
        return c_vector
```

```python
elementIndividual = ElementIndividual(1, 9, 8, 3, 5, 10, 13, 10, 5, 15, 3, 132)
class AlgoritmoGenetico():
    """
        Algoritmo genético
    """
    qtd = 0
    tam_mapa = 132 # valor máixo do tamanho do jogo
    def __init__(self, x_min, x_max, tam_populacao, taxa_mutacao, taxa_crossover, num_geracoes):
        """
            Inicializa todos os atributos da instância
        """
        self.x_min = x_min
        self.x_max = x_max
        self.tam_populacao = tam_populacao
        self.taxa_mutacao = taxa_mutacao
        self.taxa_crossover = taxa_crossover
        self.num_geracoes = num_geracoes
        self.ground_populacao = []
        self.block_populacao = []
        self.enemy_populacao = []
        self.coin_populacao = []
        self.ground_dp_vector = []
        self.block_dp_vector = []
        self.enemy_dp_vector = []
        self.coin_dp_vector = []
        self.ultima_geracao = []
        self.seed_ground_crossover = []
        self.seed_ground_mutacao = []
        self.seed_ground_valor = []
        self.seed_ground_corte = []
        self.seed_block_crossover = []
        self.seed_block_mutacao = []
        self.seed_block_valor = []
        self.seed_block_corte = []
        self.seed_enemy_crossover = []
        self.seed_enemy_mutacao = []
        self.seed_enemy_valor = []
        self.seed_enemy_corte = []
        self.seed_coin_crossover = []
        self.seed_coin_mutacao = []
        self.seed_coin_valor = []
        self.seed_coin_corte = []
        self.mapa_jogabilidade = []
        self.qtd += 1
        # gera os individuos da população

    def _gerar_populacao(self, elemento):
        """
            Gera uma população de um determinado tamanho com individuos que possuem um número
            expecífico de bits
        """
        # Ground
        if elemento == "Ground":
            for _ in range(0, self.tam_populacao):
                self.ground_populacao.append(elementIndividual._groundIndividual())
        # Block
        elif elemento == "Block":
            for _ in range(0, self.tam_populacao):
                self.block_populacao.append(elementIndividual._blockIndividual())
        # Enemy
        elif elemento == "Enemy":
            for _ in range(0, self.tam_populacao):
                self.enemy_populacao.append(elementIndividual._enemyIndividual())
        # Coin
        elif elemento == "Coin":
            for _ in range(0, self.tam_populacao):
                self.coin_populacao.append(elementIndividual._coinIndividual())
                
    def setar_populacao_inicial(self, ground_populacao, block_populacao, enemy_populacao, coin_populacao):
        self.ground_populacao = ground_populacao.copy()
        self.block_populacao = block_populacao.copy()
        self.enemy_populacao = enemy_populacao.copy()
        self.coin_populacao = coin_populacao.copy()
            
    def setar_seed(self, ground_seed, block_seed, enemy_seed, coin_seed):
        self.seed_ground_crossover = ground_seed[0]
        self.seed_ground_mutacao = ground_seed[1]
        self.seed_ground_valor = ground_seed[2]
        self.seed_ground_corte = ground_seed[3]

        self.seed_block_crossover = block_seed[0]
        self.seed_block_mutacao = block_seed[1]
        self.seed_block_valor = block_seed[2]
        self.seed_block_corte = block_seed[3]

        self.seed_enemy_crossover = enemy_seed[0]
        self.seed_enemy_mutacao = enemy_seed[1]
        self.seed_enemy_valor = enemy_seed[2]
        self.seed_enemy_corte = enemy_seed[3]

        self.seed_coin_crossover = coin_seed[0]
        self.seed_coin_mutacao = coin_seed[1]
        self.seed_coin_valor = coin_seed[2]
        self.seed_coin_corte = coin_seed[3]

    def avaliarPopulacao(self, elemento):
        """
            Avalia as souluções produzidas, associando uma nota/avalição a cada elemento da população
        """
        if elemento == "Ground":
            self.ground_populacao_dp_vector = []
            for individuo_ground in self.ground_populacao:
    #             print("População1:\n", individuo)
                varia_ground = np.std(individuo_ground)
                soma_ground = 0
                inicio_ground = individuo_ground[0]
                for i_ground in range(1, len(individuo_ground)):
                    if inicio_ground != individuo_ground[i_ground]:
                        soma_ground += 1
    #             print("Desvio Padrão*:\n", (soma/self.elementIndividual.chromossomeSize))
                self.ground_populacao_dp_vector.append((soma_ground/elementIndividual.chromossomeSize))
                
            self.ground_dp_vector.append(self.ground_populacao_dp_vector[self.ground_populacao_dp_vector.index(min(self.ground_populacao_dp_vector))])
        elif elemento == "Block":
            self.block_populacao_dp_vector = []
            for individuo_block in self.block_populacao:
    #             print("População1:\n", individuo)
                varia_block = np.std(individuo_block)
                soma_block = 0
                inicio_block = individuo_block[0]
                for i_block in range(1, len(individuo_block)):
                    if inicio_block != individuo_block[i_block]:
                        soma_block += 1
    #             print("Desvio Padrão*:\n", (soma/self.elementIndividual.chromossomeSize))
                self.block_populacao_dp_vector.append((soma_block/elementIndividual.chromossomeSize))
                
            self.block_dp_vector.append(self.block_populacao_dp_vector[self.block_populacao_dp_vector.index(min(self.block_populacao_dp_vector))])
        elif elemento == "Enemy":
            self.enemy_populacao_dp_vector = []
            for individuo_enemy in self.enemy_populacao:
    #             print("População1:\n", individuo)
                varia_enemy = np.std(individuo_enemy)
                soma_enemy = 0
                inicio_enemy = individuo_enemy[0]
                for i_enemy in range(1, len(individuo_enemy)):
                    if inicio_enemy != individuo_enemy[i_enemy]:
                        soma_enemy += 1
    #             print("Desvio Padrão*:\n", (soma/self.elementIndividual.chromossomeSize))
                self.enemy_populacao_dp_vector.append((soma_enemy/elementIndividual.chromossomeSize))
                
            self.enemy_dp_vector.append(self.enemy_populacao_dp_vector[self.enemy_populacao_dp_vector.index(min(self.enemy_populacao_dp_vector))])
        elif elemento == "Coin":
            self.coin_populacao_dp_vector = []
            for individuo_coin in self.coin_populacao:
    #             print("População1:\n", individuo)
                varia_coin = np.std(individuo_coin)
                soma_coin = 0
                inicio_coin = individuo_coin[0]
                for i_coin in range(1, len(individuo_coin)):
                    if inicio_coin != individuo_coin[i_coin]:
                        soma_coin += 1
    #             print("Desvio Padrão*:\n", (soma/self.elementIndividual.chromossomeSize))
                self.coin_populacao_dp_vector.append((soma_coin/elementIndividual.chromossomeSize))
                
            self.coin_dp_vector.append(self.coin_populacao_dp_vector[self.coin_populacao_dp_vector.index(min(self.coin_populacao_dp_vector))])
        
    def _selecionarPais(self, elemento):
        """
            Realiza a seleção do individuo mais apto por torneio, considerando N = 2
        """
        p1_1 = []
        p1_2 = []
        participantes_torneio = []
        # agrupa os individuos com suas avaliações para gerar os participantes do torneio
        if elemento == "Ground":
            index = self.ground_populacao_dp_vector.index(min(self.ground_populacao_dp_vector))
            p1_1.append(self.ground_populacao[index])
            p1_1_dp = self.ground_populacao_dp_vector[index]
            self.ground_populacao.remove(self.ground_populacao[index])
            self.ground_populacao_dp_vector.remove(self.ground_populacao_dp_vector[index])
            
            index = self.ground_populacao_dp_vector.index(min(self.ground_populacao_dp_vector))
            p1_2.append(self.ground_populacao[index])
            p1_2_dp = self.ground_populacao_dp_vector[index]
            self.ground_populacao.remove(self.ground_populacao[index])
            self.ground_populacao_dp_vector.remove(self.ground_populacao_dp_vector[index])
        elif elemento == "Block":
            index = self.block_populacao_dp_vector.index(min(self.block_populacao_dp_vector))
            p1_1.append(self.block_populacao[index])
            p1_1_dp = self.block_populacao_dp_vector[index]
            self.block_populacao.remove(self.block_populacao[index])
            self.block_populacao_dp_vector.remove(self.block_populacao_dp_vector[index])
            
            index = self.block_populacao_dp_vector.index(min(self.block_populacao_dp_vector))
            p1_2.append(self.block_populacao[index])
            p1_2_dp = self.block_populacao_dp_vector[index]
            self.block_populacao.remove(self.block_populacao[index])
            self.block_populacao_dp_vector.remove(self.block_populacao_dp_vector[index])
        elif elemento == "Enemy":
            index = self.enemy_populacao_dp_vector.index(min(self.enemy_populacao_dp_vector))
            p1_1.append(self.enemy_populacao[index])
            p1_1_dp = self.enemy_populacao_dp_vector[index]
            self.enemy_populacao.remove(self.enemy_populacao[index])
            self.enemy_populacao_dp_vector.remove(self.enemy_populacao_dp_vector[index])
            
            index = self.enemy_populacao_dp_vector.index(min(self.enemy_populacao_dp_vector))
            p1_2.append(self.enemy_populacao[index])
            p1_2_dp = self.enemy_populacao_dp_vector[index]
            self.enemy_populacao.remove(self.enemy_populacao[index])
            self.enemy_populacao_dp_vector.remove(self.enemy_populacao_dp_vector[index])
        elif elemento == "Coin":
            index = self.coin_populacao_dp_vector.index(min(self.coin_populacao_dp_vector))
            p1_1.append(self.coin_populacao[index])
            p1_1_dp = self.coin_populacao_dp_vector[index]
            self.coin_populacao.remove(self.coin_populacao[index])
            self.coin_populacao_dp_vector.remove(self.coin_populacao_dp_vector[index])
            
            index = self.coin_populacao_dp_vector.index(min(self.coin_populacao_dp_vector))
            p1_2.append(self.coin_populacao[index])
            p1_2_dp = self.coin_populacao_dp_vector[index]
            self.coin_populacao.remove(self.coin_populacao[index])
            self.coin_populacao_dp_vector.remove(self.coin_populacao_dp_vector[index])
        
        return p1_1, p1_2, p1_1_dp, p1_2_dp
        
    def onePointCrossover(self, p1_1, p1_2, elem):
        """
            Aplica o crossover de acordo com uma dada probabilidade (taxa de crossover)
        """
        p2_1 = []
        p2_2 = []
        # ground crossover
        p1_1_elem = p1_1[0]
        p1_2_elem = p1_2[0]
        if randint(1,100) <= self.taxa_crossover:
            if elem == "Ground":
                self.seed_ground_crossover.append(1)
            elif elem == "Block":
                self.seed_block_crossover.append(1)
            elif elem == "Enemy":
                self.seed_enemy_crossover.append(1)
            elif elem == "Coin":
                self.seed_coin_crossover.append(1)
            # caso o crossover seja aplicado os pais trocam suas caldas e com isso geram dois filhos
            ponto_de_corte1 = randint(1, int(self.tam_mapa)/2)
            ponto_de_corte2 = randint(int(self.tam_mapa)/2, int(self.tam_mapa))
            if elem == "Ground":
                self.seed_ground_corte.append(ponto_de_corte1)
                self.seed_ground_corte.append(ponto_de_corte2)
            elif elem == "Block":
                self.seed_block_corte.append(ponto_de_corte1)
                self.seed_block_corte.append(ponto_de_corte2)
            elif elem == "Enemy":
                self.seed_enemy_corte.append(ponto_de_corte1)
                self.seed_enemy_corte.append(ponto_de_corte2)
            elif elem == "Coin":
                self.seed_coin_corte.append(ponto_de_corte1)
                self.seed_coin_corte.append(ponto_de_corte2)
            p2_1.append(p1_1_elem[:ponto_de_corte1] + p1_2_elem[ponto_de_corte1:ponto_de_corte2] + p1_1_elem[ponto_de_corte2:])
            p2_2.append(p1_2_elem[:ponto_de_corte1] + p1_1_elem[ponto_de_corte1:ponto_de_corte2] + p1_2_elem[ponto_de_corte2:])
        else:
            if elem == "Ground":
                self.seed_ground_crossover.append(0)
            elif elem == "Block":
                self.seed_block_crossover.append(0)
            elif elem == "Enemy":
                self.seed_enemy_crossover.append(0)
            elif elem == "Coin":
                self.seed_coin_crossover.append(0)
            # caso contrário os filhos são cópias exatas dos pais
            p2_1.append(p1_1_elem[:])
            p2_2.append(p1_2_elem[:])
            
        return p2_1, p2_2
        
    def mutation(self, individuo, elem):
        """
            Realiza a mutação dos bits de um indiviuo conforme uma dada probabilidade
            (taxa de mutação)
        """
        # caso a taxa de mutação seja atingida, ela é realizada em um bit aleatório
        # ground
        individuo_elem = individuo[0]
        for i in range(0, len(individuo_elem) - 1):
            if randint(1,100) <= self.taxa_mutacao:
                if elem == "Ground":
                    self.seed_ground_mutacao.append(1)
                elif elem == "Block":
                    self.seed_block_mutacao.append(1)
                elif elem == "Enemy":
                    self.seed_enemy_mutacao.append(1)
                elif elem == "Coin":
                    self.seed_coin_mutacao.append(1)
                valor = individuo[0][i] + randint(-4, 4)
                if elem == "Ground":
                    self.seed_ground_valor.append(valor)
                elif elem == "Block":
                    self.seed_block_valor.append(valor)
                elif elem == "Enemy":
                    self.seed_enemy_valor.append(valor)
                elif elem == "Coin":
                    self.seed_coin_valor.append(valor)
                if valor < 0:
                    individuo[0][i] = 0
                elif valor > 8:
                    individuo[0][i] = 8
                else:
                    individuo[0][i] = valor
            else:
                if elem == "Ground":
                    self.seed_ground_mutacao.append(0)
                elif elem == "Block":
                    self.seed_block_mutacao.append(0)
                elif elem == "Enemy":
                    self.seed_enemy_mutacao.append(0)
                elif elem == "Coin":
                    self.seed_coin_mutacao.append(0)
                
        return individuo
    
    def selecionarMelhorIndividuo(self, elemento):
        """
            Avalia as souluções produzidas, associando uma nota/avalição a cada elemento da população
        """
        if elemento == "Ground":
            ground_ugeracao_dp_vector = []
            for individuo in self.ground_populacao:
                varia = np.std(individuo)
                soma = 0
                inicio = individuo[0]
                for i in range(1, len(individuo)):
                    if inicio != individuo[i]:
                        soma += 1
                ground_ugeracao_dp_vector.append((soma/elementIndividual.chromossomeSize))
            
            self.ultima_geracao.append(self.ground_populacao[ground_ugeracao_dp_vector.index(min(ground_ugeracao_dp_vector))])
        elif elemento == "Block":
            block_ugeracao_dp_vector = []
            for individuo in self.block_populacao:
                varia = np.std(individuo)
                soma = 0
                inicio = individuo[0]
                for i in range(1, len(individuo)):
                    if inicio != individuo[i]:
                        soma += 1
                block_ugeracao_dp_vector.append((soma/elementIndividual.chromossomeSize))
            
            self.ultima_geracao.append(self.block_populacao[block_ugeracao_dp_vector.index(min(block_ugeracao_dp_vector))])
        elif elemento == "Enemy":
            enemy_ugeracao_dp_vector = []
            for individuo in self.enemy_populacao:
                varia = np.std(individuo)
                soma = 0
                inicio = individuo[0]
                for i in range(1, len(individuo)):
                    if inicio != individuo[i]:
                        soma += 1
                enemy_ugeracao_dp_vector.append((soma/elementIndividual.chromossomeSize))
            
            self.ultima_geracao.append(self.enemy_populacao[enemy_ugeracao_dp_vector.index(min(enemy_ugeracao_dp_vector))])
        elif elemento == "Coin":
            coin_ugeracao_dp_vector = []
            for individuo in self.coin_populacao:
                varia = np.std(individuo)
                soma = 0
                inicio = individuo[0]
                for i in range(1, len(individuo)):
                    if inicio != individuo[i]:
                        soma += 1
                coin_ugeracao_dp_vector.append((soma/elementIndividual.chromossomeSize))
            
            self.ultima_geracao.append(self.coin_populacao[coin_ugeracao_dp_vector.index(min(coin_ugeracao_dp_vector))])
            
    def avaliarFilho(self, p2_1, p2_2, elem):
        """
            Avalia as souluções produzidas, associando uma nota/avalição a cada elemento da população
        """
        if elem == "Ground":
            for individuo_ground in p2_1:
    #             print("População2_1:\n", individuo)
                varia_ground = np.std(individuo_ground)
                soma_ground = 0
                inicio_ground = individuo_ground[0]
                for i_ground in range(1, len(individuo_ground)):
                    if inicio_ground != individuo_ground[i_ground]:
                        soma_ground += 1
                p2_1_dp_ground = (soma_ground/elementIndividual.chromossomeSize)
    #             print("Desvio Padrão*:\n", p2_1_dp)

            for individuo_ground in p2_2:
    #             print("População2_2:\n", individuo)
                varia_ground = np.std(individuo_ground)
                soma_ground = 0
                inicio_ground = individuo_ground[0]
                for i_ground in range(1, len(individuo_ground)):
                    if inicio_ground != individuo_ground[i_ground]:
                        soma_ground += 1
                p2_2_dp_ground = (soma_ground/elementIndividual.chromossomeSize)
    #             print("Desvio Padrão*:\n", p2_2_dp)

            return p2_1_dp_ground, p2_2_dp_ground
        elif elem == "Block":
            for individuo_block in p2_1:
    #             print("População2_1:\n", individuo)
                varia_block = np.std(individuo_block)
                soma_block = 0
                inicio_block = individuo_block[0]
                for i_block in range(1, len(individuo_block)):
                    if inicio_block != individuo_block[i_block]:
                        soma_block += 1
                p2_1_dp_block = (soma_block/elementIndividual.chromossomeSize)
    #             print("Desvio Padrão*:\n", p2_1_dp)

            for individuo_block in p2_2:
    #             print("População2_2:\n", individuo)
                varia_block = np.std(individuo_block)
                soma_block = 0
                inicio_block = individuo_block[0]
                for i_block in range(1, len(individuo_block)):
                    if inicio_block != individuo_block[i_block]:
                        soma_block += 1
                p2_2_dp_block = (soma_block/elementIndividual.chromossomeSize)
    #             print("Desvio Padrão*:\n", p2_2_dp)

            return p2_1_dp_block, p2_2_dp_block
        elif elem == "Enemy":
            for individuo_enemy in p2_1:
    #             print("População2_1:\n", individuo)
                varia_enemy = np.std(individuo_enemy)
                soma_enemy = 0
                inicio_enemy = individuo_enemy[0]
                for i_enemy in range(1, len(individuo_enemy)):
                    if inicio_enemy != individuo_enemy[i_enemy]:
                        soma_enemy += 1
                p2_1_dp_enemy = (soma_enemy/elementIndividual.chromossomeSize)
    #             print("Desvio Padrão*:\n", p2_1_dp)

            for individuo_enemy in p2_2:
    #             print("População2_2:\n", individuo)
                varia_enemy = np.std(individuo_enemy)
                soma_enemy = 0
                inicio_enemy = individuo_enemy[0]
                for i_enemy in range(1, len(individuo_enemy)):
                    if inicio_enemy != individuo_enemy[i_enemy]:
                        soma_enemy += 1
                p2_2_dp_enemy = (soma_enemy/elementIndividual.chromossomeSize)
    #             print("Desvio Padrão*:\n", p2_2_dp)

            return p2_1_dp_enemy, p2_2_dp_enemy
        elif elem == "Coin":
            for individuo_coin in p2_1:
    #             print("População2_1:\n", individuo)
                varia_coin = np.std(individuo_coin)
                soma_coin = 0
                inicio_coin = individuo_coin[0]
                for i_coin in range(1, len(individuo_coin)):
                    if inicio_coin != individuo_coin[i_coin]:
                        soma_coin += 1
                p2_1_dp_coin = (soma_coin/elementIndividual.chromossomeSize)
    #             print("Desvio Padrão*:\n", p2_1_dp)

            for individuo_coin in p2_2:
    #             print("População2_2:\n", individuo)
                varia_coin = np.std(individuo_coin)
                soma_coin = 0
                inicio_coin = individuo_coin[0]
                for i_coin in range(1, len(individuo_coin)):
                    if inicio_coin != individuo_coin[i_coin]:
                        soma_coin += 1
                p2_2_dp_coin = (soma_coin/elementIndividual.chromossomeSize)
    #             print("Desvio Padrão*:\n", p2_2_dp)

            return p2_1_dp_coin, p2_2_dp_coin
    
    def gerar_mapa(self, tipo):
        ng_ground = self.ultima_geracao[0]
        ng_ground_aux = self.ultima_geracao[0]
        ng_block = self.ultima_geracao[1]
        ng_enemy = self.ultima_geracao[2]
        ng_coin = self.ultima_geracao[3]
        ng_block_height = []
        ng_enemy_height = []
        ng_coin_height = []
        for height in ng_ground:
            ng_block_height.append(height + elementIndividual.min_height_block)
            ng_enemy_height.append(height + 1)
            ng_coin_height.append(height + elementIndividual.max_coin_height)

        lines = []
        column = ""
        end = 0
        
        for j in range(0, 16):
            for i in range(0, self.tam_mapa - 1):
                if ng_ground[i] == 1 and j == 0:
                    column = column + "X"
                    ng_ground[i] = ng_ground[i] - 1
                elif ng_ground[i] > 1:
                    column = column + "|"
                    ng_ground[i] = ng_ground[i] - 1
                elif ng_ground[i] == 1:
                    column = column + "%"
                    ng_ground[i] = ng_ground[i] - 1
                else:
                    if j + 1 == ng_enemy_height[i]:
                        column = column + elementIndividual.e[ng_enemy[i]]
                    elif j + 1 == ng_block_height[i]:
                        column = column + elementIndividual.b[ng_block[i]]
                    elif j + 1 == ng_coin_height[i]:
                        column = column + elementIndividual.c[ng_coin[i]]
                    else:
                        column = column + "-"
            lines.append(column)
            column = ""
        
        if tipo == "Sequencial":
            f = open("C:/Users/rapha/Desktop/Mario-AI-Framework-master/levels/new_level/seq_level_randomized.txt", "w+")
#             f = open("C:/Users/rapha/Desktop/algoritmo_genetico-master/niveis/seq_level_randomized.txt", "w+")
        elif tipo == "Paralelo":
            f = open("C:/Users/rapha/Desktop/Mario-AI-Framework-master/levels/new_level/par_level_randomized.txt", "w+")
#             f = open("C:/Users/rapha/Desktop/algoritmo_genetico-master/niveis/par_level_randomized.txt", "w+")

        for j in range(0, 16):
            f.write("" + lines[15 - j] + "\n")

        f.close()
        
    def onePointCrossoverSeed(self, p1_1, p1_2, elem):
        """
            Aplica o crossover de acordo com uma dada probabilidade (taxa de crossover)
        """
        p2_1 = []
        p2_2 = []
        # ground crossover
        p1_1_elem = p1_1[0]
        p1_2_elem = p1_2[0]
        
        if elem == "Ground":
            seed_atual = self.seed_ground_crossover[0]
            self.seed_ground_crossover.pop(0)
            if seed_atual == 1:
                seed_corte_atual1 = self.seed_ground_corte[0]
                self.seed_ground_corte.pop(0)
                seed_corte_atual2 = self.seed_ground_corte[0]
                self.seed_ground_corte.pop(0)
                # caso o crossover seja aplicado os pais trocam suas caldas e com isso geram dois filhos
                ponto_de_corte1 = seed_corte_atual1
                ponto_de_corte2 = seed_corte_atual2
                p2_1.append(p1_1_elem[:ponto_de_corte1] + p1_2_elem[ponto_de_corte1:ponto_de_corte2] + p1_1_elem[ponto_de_corte2:])
                p2_2.append(p1_2_elem[:ponto_de_corte1] + p1_1_elem[ponto_de_corte1:ponto_de_corte2] + p1_2_elem[ponto_de_corte2:])
            else:
                # caso contrário os filhos são cópias exatas dos pais
                p2_1.append(p1_1_elem[:])
                p2_2.append(p1_2_elem[:])
        elif elem == "Block":
            seed_atual = self.seed_block_crossover[0]
            self.seed_block_crossover.pop(0)
            if seed_atual == 1:
                seed_corte_atual1 = self.seed_block_corte[0]
                self.seed_block_corte.pop(0)
                seed_corte_atual2 = self.seed_block_corte[0]
                self.seed_block_corte.pop(0)
                # caso o crossover seja aplicado os pais trocam suas caldas e com isso geram dois filhos
                ponto_de_corte1 = seed_corte_atual1
                ponto_de_corte2 = seed_corte_atual2
                p2_1.append(p1_1_elem[:ponto_de_corte1] + p1_2_elem[ponto_de_corte1:ponto_de_corte2] + p1_1_elem[ponto_de_corte2:])
                p2_2.append(p1_2_elem[:ponto_de_corte1] + p1_1_elem[ponto_de_corte1:ponto_de_corte2] + p1_2_elem[ponto_de_corte2:])
            else:
                # caso contrário os filhos são cópias exatas dos pais
                p2_1.append(p1_1_elem[:])
                p2_2.append(p1_2_elem[:])
        elif elem == "Enemy":
            seed_atual = self.seed_enemy_crossover[0]
            self.seed_enemy_crossover.pop(0)
            if seed_atual == 1:
                seed_corte_atual1 = self.seed_enemy_corte[0]
                self.seed_enemy_corte.pop(0)
                seed_corte_atual2 = self.seed_enemy_corte[0]
                self.seed_enemy_corte.pop(0)
                # caso o crossover seja aplicado os pais trocam suas caldas e com isso geram dois filhos
                ponto_de_corte1 = seed_corte_atual1
                ponto_de_corte2 = seed_corte_atual2
                p2_1.append(p1_1_elem[:ponto_de_corte1] + p1_2_elem[ponto_de_corte1:ponto_de_corte2] + p1_1_elem[ponto_de_corte2:])
                p2_2.append(p1_2_elem[:ponto_de_corte1] + p1_1_elem[ponto_de_corte1:ponto_de_corte2] + p1_2_elem[ponto_de_corte2:])
            else:
                # caso contrário os filhos são cópias exatas dos pais
                p2_1.append(p1_1_elem[:])
                p2_2.append(p1_2_elem[:])
        elif elem == "Coin":
            seed_atual = self.seed_coin_crossover[0]
            self.seed_coin_crossover.pop(0)
            if seed_atual == 1:
                seed_corte_atual1 = self.seed_coin_corte[0]
                self.seed_coin_corte.pop(0)
                seed_corte_atual2 = self.seed_coin_corte[0]
                self.seed_coin_corte.pop(0)
                # caso o crossover seja aplicado os pais trocam suas caldas e com isso geram dois filhos
                ponto_de_corte1 = seed_corte_atual1
                ponto_de_corte2 = seed_corte_atual2
                p2_1.append(p1_1_elem[:ponto_de_corte1] + p1_2_elem[ponto_de_corte1:ponto_de_corte2] + p1_1_elem[ponto_de_corte2:])
                p2_2.append(p1_2_elem[:ponto_de_corte1] + p1_1_elem[ponto_de_corte1:ponto_de_corte2] + p1_2_elem[ponto_de_corte2:])
            else:
                # caso contrário os filhos são cópias exatas dos pais
                p2_1.append(p1_1_elem[:])
                p2_2.append(p1_2_elem[:])
            
        return p2_1, p2_2
        
    def mutationSeed(self, individuo, elem):
        """
            Realiza a mutação dos bits de um indiviuo conforme uma dada probabilidade
            (taxa de mutação)
        """
        # caso a taxa de mutação seja atingida, ela é realizada em um bit aleatório
        # ground
        individuo_elem = individuo[0]
        for i in range(0, len(individuo_elem) - 1):
            if elem == "Ground":
                seed_atual = self.seed_ground_mutacao[0]
                self.seed_ground_mutacao.remove(self.seed_ground_mutacao[0])
            elif elem == "Block":
                seed_atual = self.seed_block_mutacao[0]
                self.seed_block_mutacao.remove(self.seed_block_mutacao[0])
            elif elem == "Enemy":
                seed_atual = self.seed_enemy_mutacao[0]
                self.seed_enemy_mutacao.remove(self.seed_enemy_mutacao[0])
            elif elem == "Coin":
                seed_atual = self.seed_coin_mutacao[0]
                self.seed_coin_mutacao.remove(self.seed_coin_mutacao[0])
            if seed_atual == 1:
                if elem == "Ground":
                    valor = self.seed_ground_valor[0]
                    self.seed_ground_valor.remove(self.seed_ground_valor[0])
                elif elem == "Block":
                    valor = self.seed_block_valor[0]
                    self.seed_block_valor.remove(self.seed_block_valor[0])
                elif elem == "Enemy":
                    valor = self.seed_enemy_valor[0]
                    self.seed_enemy_valor.remove(self.seed_enemy_valor[0])
                elif elem == "Coin":
                    valor = self.seed_coin_valor[0]
                    self.seed_coin_valor.remove(self.seed_coin_valor[0])
                if valor < 0:
                    individuo[0][i] = 0
                elif valor > 8:
                    individuo[0][i] = 8
                else:
                    individuo[0][i] = valor
                
        return individuo
    
    def avaliarJogabilidade(self):
        for j in range(self.tam_populacao):
            if self.ground_populacao[j][0] == 0 or self.ground_populacao[j][len(self.ground_populacao[j]) - 1] == 0:
                jogabilidade = 0
            else:
                for i in range(len(self.ground_populacao[j]) - 9):
                    if self.ground_populacao[j][i + 1] == 0:
                        if self.ground_populacao[j][i + 9] == 0 and i + 9 < len(self.ground_populacao[j]):
                            if self.block_populacao[j][i + 1] == 0:
                                jogabilidade = 0
                            else:
                                jogabilidade = 1
                        else:
                            jogabilidade = 1
                    else:
                        jogabilidade = 1
            self.mapa_jogabilidade.append(jogabilidade)
```
